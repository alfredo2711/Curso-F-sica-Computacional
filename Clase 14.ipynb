# Ejemplo sencillo de diferencia finita

Encontrar la solución de la ecuación:
$\frac{dy}{dx}=sin(5x)$, para $x=0$ $y=2$
Utilizar aproximación forward

Comparar con la solución analítica
$y=-\frac{cos(5x)}{5} + \frac{11}{5}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h=0.001
lim_inf=0
lim_sup=10
N=int((lim_sup-lim_inf)/h)
```

```python
N
```

```python
x=np.zeros(N)
y=np.zeros(N)
```

```python
#condición inicial
y[0]=2
x[0]=0
```

discretizando la derivada con la aproximación forward tenemos:
$\frac{dy}{dx}=\frac{y_{i+1}-y_{i}}{h}$

Al final podemos escribir:
$\frac{y_{i+1}-y_{i}}{h}=sin(5x_{i})$

Despejando $y_{i+1}$, obtenemos

$y_{i+1}=y_{i}+hsin(5x_{i})$

```python
for i in range(0,N-1):
    y[i+1]=y[i]+h*np.sin(5*x[i])
    x[i+1]=x[i]+h
```

```python
plt.plot(x,y)
```

```python
#solución analítica
f=-np.cos(5*x)/5 + 11/5
```

```python
plt.plot(x, y, 'r', label='numérica')
plt.plot(x,f,'b', label='analítica')
plt.legend()
```

```python
plt.plot(x,f-y)
```

```python
#cálculo del número pi
x=np.random.random(1000)
y=np.random.random(1000)                  
```

```python
ángulo=np.linspace(0,2*np.pi,100)
```

```python
fig, axes=plt.subplots(1,1, figsize=(6,6))
axes.scatter(x,y)
axes.plot(np.cos(ángulo),np.sin(ángulo),'r')
axes.set_ylim(0,1)
axes.set_xlim(0,1)
```
