## Decaimiento Radiactivo

El proceso de radiaci√≥n fue observado por primera vez por Henri Becquerel en 1896, y la palabra radiactivo (inicialmente radioactivo) fue acu√±ada por Marie Curie en 1898 para describir materiales en los que los n√∫cleos, o partes centrales de los √°tomos, son inestables y pueden cambiar espont√°neamente

```python
#Imagen de Henri y Curie
display.display(Image.open('descubridores.png'))
```

Debido a las cargas positivas de los protones estos est√°n constantemente rechaz√°ndose los unos a los otros.  El n√∫cleo se mantiene unido por una contrafuerza que se denomina interacci√≥n nuclear fuerte generada por los protones y neutrones. La mayor√≠a se mantiene unida en una situaci√≥n estable, pero algunas combinaciones de protones y neutrones no pueden mantener el equilibrio y espont√°neamente (debido a una interacci√≥n llamada d√©bil) cambian o decaen hasta alcanzar una forma m√°s estable liberando part√≠culas y/o radiaci√≥n electromagn√©tica. 
Estos cambios son conocidos como radiactividad nuclear, y esos n√∫cleos son is√≥topos radiactivos o radionucleidos.  Para alcanzar su forma m√°s estable algo de la energ√≠a del n√∫cleo debe ser liberada y esta energ√≠a que se libera es la que se conoce como radiaci√≥n.

## Desintegraci√≥n Beta 
Seg√∫n el nucleido, si tiene exceso o defecto de neutrones, tratar√° de ganar protones en su n√∫cleo o de perderlos para lograr la estabilidad. Esto da lugar a los procesos de emisi√≥n de part√≠culas beta negativas (Œ≤-) o beta positivas (Œ≤+) respectivamente. 

### Desintegraci√≥n Beta negativa
Cuando la relaci√≥n de neutrones a protones es mayor que la correspondiente a la estabilidad, el nucleido se puede desintegrar por medio de un proceso equivalente a la transformaci√≥n de un neutr√≥n en un prot√≥n, emiti√©ndose una part√≠cula Œ≤ negativa y un (anti)neutrino.

La ecuaci√≥n general es: $$ _{Z}^{A}\textrm{X}\rightarrow  _{Z+1}^{A}\textrm{Y} +\beta ^{-}+\nu $$

Donde $ v $ es un (anti)neutrino

### Desintegraci√≥n Beta positiva
En la naturaleza no existen emisores Œ≤+, sino que se producen por acci√≥n de deuterones (n√∫cleos de deuterio) o part√≠culas Œ± que inciden sobre los √°tomos de ciertos elementos.
La desintegraci√≥n beta positiva consiste en un proceso equivalente a la transformaci√≥n de un prot√≥n en un neutr√≥n, emiti√©ndose un positr√≥n y un neutrino. 

La ecuaci√≥n general es: $$ _{Z}^{A}\textrm{X}\rightarrow _{Z-1}^{A}\textrm{Y} +\beta ^{+}+\nu $$

Donde $ v $ es un neutrino

## Por Captura Electr√≥nica
Los n√∫cleos con relaci√≥n neutr√≥n a prot√≥n en defecto, pueden pasar a una condici√≥n estable por captura electr√≥nica. En este caso se captura un electr√≥n orbital (b√°sicamente de la capa K), posteriormente, tiene lugar un proceso equivalente a la transformaci√≥n de un prot√≥n en un neutr√≥n emiti√©ndose s√≥lo un neutrino. 

La ecuaci√≥n general es la siguiente: $$ e^{-}+_{A}^{Z}\textrm{X}\rightarrow _{Z-1}^{A}\textrm{Y}+\nu  $$


## Radiaci√≥n Gamma
Como consecuencia de un proceso de desintegraci√≥n Œ± o Œ≤ o de captura electr√≥nica el nucleido hijo puede quedar en un nivel fundamental o bien un estado excitado, entendi√©ndose por tal, aquel en el que posee un exceso de energ√≠a con respecto al nivel fundamental. En estos casos el n√∫cleo se desexcita emitiendo radiaci√≥n electromagn√©tica. Esta radiaci√≥n conocida como gamma es similar a la luz visible o a los rayos X pero su origen es nuclear y por tanto su frecuencia es m√°s elevada. 
 
 La ecuaci√≥n General: $$ _{Z}^{A}\textrm{X}^{*}\rightarrow _{Z}^{A}\textrm{X} +\gamma $$
 
 Donde $ \gamma $ es un fot√≥n 
 

## Decaimiento radiactivo

Una manera de ver las leyes que rigen el decaimiento radiactivo es utilizando la curva experimental del $ _{}^{14}\textrm{C} $

```python
#Imagen de decaimiento del C-14
display.display(Image.open('decaimiento.png'))
```

La expresi√≥n que describe la ley exponencial del decaimiento es la siguiente $$ N= N_{0} e^{-\lambda t} $$ 


En esta ecuaci√≥n, N representa el n√∫mero de n√∫cleos del padre en el instante t; $ N_{0} $ es el n√∫mero de n√∫cleos iniciales cuando t = 0; Œª se denomina la constante de decaimiento; e es la base de los logaritmos naturales. La constante de decaimiento Œª representa la probabilidad de que haya decaimiento 

Implementamos el concepto de periodo de semidecintegraci√≥n $ t_{\frac{1}{2}} $, donde por cada periodo $N(t_{\frac{1}{2}}) \rightarrow \frac{N_{0}}{2}$ por sustituci√≥n nos queda que $$ t_{\frac{1}{2}}=\frac{ln2}{\lambda} $$

Otro par√°metro que puede obtenerse a partir de la primera expresi√≥n es la vida media ($ \tau $) del nucleido que representa el tiempo promedio que ‚Äúviven‚Äù los √°tomos de una fuente radiactiva. $$ \tau=\frac{1}{\lambda}= \frac{t_{\frac{1}{2}} }{ln2} $$

Entonces tenemos que: 
En una sustancia radiactiva dada, el n√∫mero de √°tomos que se transforman, (-dN), en un lapso de tiempo (dt) es proporcional al n√∫mero de √°tomos presentes en ese momento (N) y con la constante de proporcionalidad Œª que es caracter√≠stica de cada nucleido se puede plantear la ecuaci√≥n siguiente: $$ \frac{dN}{dt}=\lambda N $$

El primer t√©rmino representa la rapidez con que se desintegran los √°tomos del material que antes fue definida como actividad (A) de un radiois√≥topo, luego: $$ A= \lambda N $$

se puede calcular la disminuci√≥n de la actividad en funci√≥n del tiempo: $$ A(t)= A_{0}e^{-\lambda \tau } $$

```python
#Simulamos el decaimiento radiactivo con matplotlib
import matplotlib.pyplot as plt
```

```python
#definimos variables
p = 1/5730
dt= 1
```

```python
# Establecemos un bucle
n = 10000
t = 0.0

tlist =[]
nlist =[]

while t < 20000:
    n += -n*p*dt
    t += dt
    nlist.append(n)
    tlist.append(t)
```

```python
# graficamos
plt.plot(tlist,nlist,"-r")
plt.xlabel("Time(a)")
plt.ylabel("N")
plt.grid()
fig= plt.show()
```

```python
#simulamos con n√∫mero aleatorio y numpy
import random
import numpy as np
#definimos la funci√≥n de decaimiento
def decay_sim(thalf, N0=10000, tgrid=None, nhalflives=4):    
    tau = thalf / np.log(2)
    if tgrid is None:
        Nt, tmax = 100, thalf * nhalflives
        tgrid, dt = np.linspace(0, tmax, Nt, retstep=True)
    
    else:
        Nt = len(tgrid)
        dt = tgrid[1] - tgrid[0]
        
    N = np.empty(Nt, dtype=int)
    N[0] = N0
    #La probabilidad que tiene el nucleo en decaer
    p = dt / tau
    #aplicamos un for para incluir los n√∫meros aleatoreos
    for i in range(1, Nt):
        N[i] = N[i-1]
        for j in range(N[i-1]):
            r = random.random()
            if r < p:
                N[i] -= 1
    return tgrid, N

#Simulamos el decaimiento
N0 = 10000
thalf = 5730
Nt, tmax = 100, 20000
nsims = 1
tgrid = np.linspace(0, tmax, Nt)
Nsim = np.empty((Nt, nsims))
for i in range(nsims):
    _, Nsim[:, i] = decay_sim(thalf, N0, tgrid)
```

```python
#Graficamos la simulaci√≥n
plt.plot(tgrid, Nsim,"-r")
plt.xlabel("Time (a)")
plt.ylabel("N")
plt.plot(5730,5000, marker="o", color="black")
plt.grid()
fig = plt.show() 
```

```python
#Comparamos gr√°ficos
plt.plot(tgrid, Nsim,"-r")
plt.plot(tlist,nlist,"-b")
plt.xlabel("Time (a)")
plt.ylabel("N")
plt.grid()
fig = plt.show()
```

```python
#Animaci√≥n
from matplotlib.animation import FuncAnimation
%matplotlib notebook

x = tgrid
y = Nsim
fig = plt.figure()
plt.xlim(0, 20000)
plt.ylim(0, 10000)
graph, = plt.plot([], [], 'o')

def animate(i):
    graph.set_data(x[:i+1], y[:i+1])
    return graph

ani = FuncAnimation(fig, animate, frames=100, interval=200)
plt.xlabel("Time (a)")
plt.ylabel("N")
plt.show
```
