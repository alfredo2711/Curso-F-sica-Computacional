# Ecuaciones de Maxwell en interfaces planas 


$$\mathbf{\tilde{E}}_{I(z,t)}=\tilde{E}_{0_{I}}e^{i(k_{1z}-wt)}\mathbf{\hat{x}}\tag{1}$$        

$\mathbf{\tilde{B}}_{I(z,t)}=\frac{1}{v_{1}}\tilde{E}_{0_{I}}e^{i(k_{1z}-wt)}\mathbf{\hat{y}}\tag{2}$

$\mathbf{\tilde{E}}_{R(z,t)}=\tilde{E}_{0_{R}}e^{i(-k_{1z}-wt)}\mathbf{\hat{x}}\tag{3}$ 

$\mathbf{\tilde{B}}_{R(z,t)}=-\frac{1}{v_{1}}\tilde{E}_{0_{R}}e^{i(-k_{1z}-wt)}\mathbf{\hat{y}}\tag{4}$ 

Las ecuaciones 1,2,3 y 4 presentan un vector de onda $k_{1}$ que indica que su propagación es en el medio 1 

$\mathbf{\tilde{E}}_{T(z,t)}=\tilde{E}_{0_{T}}e^{i(k_{2z}-wt)}\mathbf{\hat{x}}\tag{5}$

$\mathbf{\tilde{B}}_{T(z,t)}=\frac{1}{v_{2}}\tilde{E}_{0_{T}}e^{i(k_{2z}-wt)}\mathbf{\hat{y}}\tag{6}$

Como vemos en las ecuaciones 5 y 6 los campos presentan un vector de onda $k_{2}$ que indica su  propagación en el medio 2 


Debido a las condiciones de frontera se requiere:


$\tilde{E}_{{0}_{I}} + \tilde{E}_{{0}_{R}} = \tilde {E}_{{0}_{T}}\tag{7}$ 


CONDICIÓN DE FRONTERA: $\frac{1}{\mu_{1}}B_1''=\frac{1}{\mu_{2}}B_{2}''$

$$\frac{1}{\mu_{1}}\left(\frac{1}{v_{1}}\tilde{E}_{{0}_{I}} - \frac{1}{v_{1}}\tilde{E}_{{0}_{R}}\right)=\frac{1}{\mu_{2}}\left(\frac{1}{v_{2}}\tilde{E}_{{0}_{T}}\right)$$                                                                               



$$\tilde{E}_{{0}_{I}} - \tilde{E}_{{0}_{R}} = \beta\tilde {E}_{{0}_{T}}\tag{8}$$

Donde Beta se expresa en la ecuación 9

$$\beta\equiv\frac{\mu_{1}v_{1}}{\mu_{2}v_{2}} = \frac{\mu_{1}n_{2}}{\mu_{2}n_{1}}\tag{9}$$

Expresando los campos reflejados y transmitidos en función del campo incidente obtenemos: 


$$ \tilde{E}_{{0}_{R}} =\left(\frac{1-\beta}{1+\beta}\right)\tilde{E}_{{0}_{I}}$$ ,
$$\tilde{E}_{{0}_{T}} =\left(\frac{2}{1+\beta}\right)\tilde{E}_{{0}_{I}}$$

$$\beta=\frac{v_{1}}{v_{2}}$$

$$\tilde{E}_{{0}_{R}}=\left(\frac{v_{2}-v_{1}}{v_{2}+v_{1}}\right)\tilde{E}_{{0}_{1}\tag{10}}$$       $$\tilde{E}_{{0}_{T}}=\left(\frac{2v_{2}}{v_{2}+v_{1}}\right)\tilde{E}_{{0}_{1}}\tag{11}$$

La intensidad promedio es proporcional al campo indicente al cuadrado. 

$$I=\frac{1}{2}\epsilon v E^{2}_{0}\tag{12}$$

Asumiendo un material no magnético:

$$R\equiv\frac{I_{R}}{I_{I}}=\left( \frac{E_{{0}_{R}}}{E_{{0}_{I}}}\right)^2=\left(\frac{n_{1}-n_{2}}{n_{1}+n_{2}}\right)^2\tag{13}$$

$$T\equiv\frac{I_{T}}{I_{I}}=\left(\frac{\epsilon_{2}v_{2}}{\epsilon_{1}v_{1}}\right)\left(\frac{E_{{0}_{T}}}{E_{{0}_{I}}}\right)^2=\frac{4n_{1}n_{2}}{(n_{1}+n_{2})^2}\tag{14}$$

Ahora consideraremos el caso de la incidencia oblicua: 

Para esta situación consideraremos dos tipos de polarizaciones incidentes la tipo s con el campo eléctrico perpendicular al plano de incidencia y la tipo p con el campo eléctrico paralelo al plano de incidencia, como se muestra en la siguiente figura. 

Siguiendo un procedimiento similar al visto anteriormente para una incidencia normal se obtienen las siguientes ecuaciones: 

Para la polarización tipo s los coeficientes de reflexión y de transmisión son: 

$$rs=\frac{E_{0R}}{E_{0I}}=\frac{2n_{1} cos \theta_{i}}{n_{1} cos \theta_{i}+\frac{\mu_{1}}{\mu_{2}}\sqrt{n_{2}^2-n_{1}^2sin^2\theta_{i}}}\tag{15}$$

$$ts=\frac{E_{0T}}{E_{0I}}=\frac{n_{1} cos \theta_{i}-\frac{\mu_{1}}{\mu_{2}}\sqrt{n_{2}^2-n_{1}^2sin^2\theta_{i}}}{n_{1} cos \theta_{i}\frac{\mu_{1}}{\mu_{2}}\sqrt{n_{2}^2-n_{1}^2sin^2\theta_{i}}}\tag{16}$$

Para la polarización tipo p los coeficientes de reflexión y transmisión son: 

$$rp=\frac{E_{OR}}{E_{0I}}=\frac{2n_{1}n_{2}cos \theta_{i}}{\frac{\mu_{1}}{\mu_{2}}n_{2}^2cos\theta_{i}+n_{1}\sqrt{n_{2}^2-n_{1}^2sin^2\theta_{i}}}\tag{17}$$

$$tp=\frac{E_{OT}}{E_{0I}}=\frac{\frac{\mu_{1}}{\mu_{2}}n_2^2cos\theta_{i}-n_{1}\sqrt{n_{2}^2-n_{1}^2sin^2\theta_{i}}}{\frac{\mu_{1}}{\mu_{2}}n_{2}^2cos\theta_{i}+n_{1}\sqrt{n_{2}^2-n_{1}^2sin^2\theta_{i}}}\tag{18}$$

Donde n2 es el índice de refracción del medio donde se transmite la onda y n1 es el índice de refracción del medio donde se da la incidencia y la reflexión

La intensidad reflejada entre la incidente (Reflectancia) se calcula:

$$Pr=|r|^2\tag{19}$$

La intensidad transmitida entre la incidente (Transmitancia) se calcula:

$$Pt=\frac{n_{2}\cos{\theta_{t}}}{n_{1}\cos{\theta_{i}}}|t|^{2}\tag{20}$$

# Códigos

```python
import numpy as np 
import matplotlib.pyplot as plt 
from numpy.lib.scimath import sqrt 
```

```python
theta=np.linspace(0, np.pi/2, 1000)
```

## De aire a vidrio

```python
n=1
n2=1.5
```

### Polarización tipo s

```python
#reflexión en polarización tipo s
rs=(n*np.cos(theta)-sqrt(n2**2-n**2*np.sin(theta)**2))/(n*np.cos(theta)+sqrt(n2**2-n**2*np.sin(theta)**2))
#transmisión en polarización tipo s (campo perpendicular al plano de incidencia)
ts=2*n*np.cos(theta)/(n*np.cos(theta)+sqrt(n2**2-n**2*np.sin(theta)**2))
#Cálculo de la Potencia reflejada
Prs=np.abs(rs)**2
#Cálculo de la Potencia transmitida
Pts=np.abs(ts)**2*(n2/n)*(np.cos(np.arcsin((n/n2)*np.sin(theta)))/np.cos(theta))
#Pts=np.abs(ts)**2*(n2/n)*np.real(sqrt(n2**2-n**2*np.sin(theta)**2))/np.cos(theta)
```

```python
fig, axes = plt.subplots(3,1,figsize=(8,10))
#Gráfica reflexión y transmisión en polarización tipo s
axes[0].plot(theta*180/np.pi,np.abs(rs), label='Reflexión',color='magenta')
axes[0].plot(theta*180/np.pi,np.abs(ts), label='Transmisión',color='blue')
axes[0].set_xlabel('ángulo incidente (°)')
#Gráfica de la potencia reflejada y transmitida en polarización tipo s
axes[1].plot(theta*180/np.pi,Prs, label='Potencia Reflejada/incidente',color='magenta')
axes[1].plot(theta*180/np.pi,Pts, label='Potencia Transmitida/insidente',color='blue')
axes[1].set_xlabel('ángulo incidente (°)')
#Gráfica de la fase reflejada y transmitida en polarización tipo s
axes[2].plot(theta*180/np.pi,np.angle(rs)*180/np.pi, label='Fase Reflejada',color='magenta')
axes[2].plot(theta*180/np.pi,np.angle(ts)*180/np.pi, label='Fase Transmitida',color='blue')
axes[2].set_xlabel('ángulo incidente (°)')
axes[0].grid()
axes[0].legend()
axes[1].grid()
axes[1].legend()
axes[2].grid()
axes[2].legend()
```

### Polarización tipo p

```python
#Reflexión en polarización tipo p 
rp=(n2**2*np.cos(theta)-n*sqrt(n2**2-n**2*np.sin(theta)**2))/(n2**2*np.cos(theta)+n*sqrt(n2**2-n**2*np.sin(theta)**2))
#Transmisión en polarización tipo p (campo perpandicular al plano de incidencia)
tp=2*n*n2*np.cos(theta)/(n2**2*np.cos(theta)+n*sqrt(n2**2-n**2*np.sin(theta)**2))
#Cálculo de la Potencia reflejada
Prp=np.abs(rp)**2
#Cálculo de la Potencia transmitida
Ptp=np.abs(tp)**2*(n2/n)*(np.cos(np.arcsin((n/n2)*np.sin(theta)))/np.cos(theta))
#Ptp=np.abs(tp)**2*(n2/n)*np.real(sqrt(n2**2-n**2*np.sin(theta)**2))/np.cos(theta)
```

```python
fig, axes = plt.subplots(3,1,figsize=(8,10))
#Gráfica de la reflexión y transmisión en polarización tipo p
axes[0].plot(theta*180/np.pi,np.abs(rp), label='Reflexión',color='magenta')
axes[0].plot(theta*180/np.pi,np.abs(tp), label='Transmisión',color='blue')
axes[0].set_xlabel('ángulo incidente (°)')
#Gráfica de la Potencia reflejada y transmitida en polarización tipo p
axes[1].plot(theta*180/np.pi,Prp, label='Potencia Reflejada/incidente',color='magenta')
axes[1].plot(theta*180/np.pi,Ptp, label='Potencia Transmitida/incidente',color='blue')
axes[1].set_xlabel('ángulo incidente (°)')
#Gráfica de la fase reflejada y transmitida en polarización tipo p
axes[2].plot(theta*180/np.pi,np.angle(rp)*180/np.pi, label='Fase Reflejada',color='magenta')
axes[2].plot(theta*180/np.pi,np.angle(tp)*180/np.pi, label='Fase Transmitida',color='blue')
axes[2].set_xlabel('ángulo incidente (°)')
axes[0].grid()
axes[0].legend()
axes[1].grid()
axes[1].legend()
axes[2].grid()
axes[2].legend()
```
