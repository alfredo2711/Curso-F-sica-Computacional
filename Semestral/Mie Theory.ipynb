## Mie Background

1.Mie's publication in 1908 is a rigerous solution of Maxwell's equations for the scattering of a linearly polarized electromagnetic plane wave $\lambda$ by a homogenous sphere of radius $r$ and with a complex refractive index $N$.

2.The power of his theory comes from it's validity for any value of particle diameter (relative to scattered light wavelength) and particle refractive index.

3.There is no simple analytical expression and must be solved numerically, hopefully, with the help of computers.

### Approximate Field Beams
Of importance to the disccussion of General Lorenz-Mie Theory (GLMT) are the various approximations of beam fields necessary for input of GLMT calculations. The Fundemental $TEM_{00}$ mode of a laser beam is of great practical utility.

In it's simplest approximation: the electric and magentic fields of a monochromatic Guassian beam of wavelength $\lambda$, wavenumber $k=\frac{2\pi}{\lambda}$, peak electric field strength $E_{0}$, Electric field half-width $w_0$,linearly polariazed in the x direaction, and traveling in the +z direction resemble a propagating Guassian-shaped tube of constant width,

$$E(x,y,z) = E_{0} exp(\frac{-\rho^2}{w_{0}^2})exp(ikz)u_{x}$$

$$B(x,y,z) = \frac{E_{0}}{c} exp(\frac{-\rho^2}{w_{0}^2})exp(ikz)u_{y}$$

Where c is the speed of light, $\rho^2 = x^2 + y^2$
and time dependence $exp(iwt)$ is implicit

If we want to approxmiate diffraction of beam field, assuming the fields in z=0 plane are Fresnel diffracted to z>0, then we get 

$$E(x,y,z) = D E_{0} exp(\frac{-D\rho^2}{w_{0}^2})exp(ikz)u_{x}$$

$$B(x,y,z) = \frac{D E_{0}}{c} exp(\frac{-D\rho^2}{w_{0}^2})exp(ikz)u_{y}$$

Where beam confiment paramter is defined as $$s=1/kw_0$$
and $$D=(1+2isz/w_0)^{-1}$$

This diffracting Gaussian beam, called the $L^{-}$ approximation to the exact $TEM_{00}$ laser beam satisfies the paraxial wave equation but does not satisfy full wave equation or Maxwell's equations. A proceedure developed to obtain the beam as a form of an infinite series in powers of s can be used to obtain what is called the L approximation but it does not extend to other beams of experimental interest such as Guassian beams with abberations after passing through a lens or "top-hat beams"

### Generalized Lorenz-Mie Theory

Consider an arbitrary transversely localized beam incident on a homogeneous sphere of radius a and refractive index N whose center is at the origin of coordinates. Aussiming the exact electric and magnetic fields of the incident beam are known and can be represented in spherical coordinates by the folloing TE and TM bromwich potentials.

$$U_{TE}(r,\theta,\phi) = \sum_{n=1}^\infty \sum_{m=-n}^n (\frac{E_0}{k})i^nC^{pw}_{n}g^{m}_{n,TE}\psi_{n}(kr)P^{\lvert{m}\rvert}_{n}[cos(\theta)]e^{im\phi}$$

$$U_{TM}(r,\theta,\phi) = \sum_{n=1}^\infty \sum_{m=-n}^n (\frac{E_0}{k})i^nC^{pw}_{n}g^{m}_{n,TM}\psi_{n}(kr)P^{\lvert{m}\rvert}_{n}[cos(\theta)]e^{im\phi}$$

Where $C^{pw}_{n} = (2n+1)/[2n(n+1)]$

$\psi_{n}$ are Riccati-Bessel functions of the form $x^2\frac{d^2y}{dx^2}+x\frac{dy}{dx}+(x^2 - \alpha^2)y = 0$ 

and 

$P^{\lvert{m}\rvert}_{n}$ are associated Legendre functions of the form $(1-x^2)\frac{d^2y}{dx^2}+2x\frac{dy}{dx}+[\lambda(\lambda+1)-\frac{\mu^2}{1-x^2}]y = 0$

and $g^{m}_{n,TE}$ and $g^{m}_{n,TM}$ are the beam shape coefficents. The calculation of these coefficents is actaully quite difficult and are obtained by inverting the equations for $U_{TE}$ and $U_{TM}$ which gives the following:

$$g^{m}_{n,TE} = \frac{(-i)^{n-1}}{2\pi}\frac{kr}{j_{n}(kr)}\frac{n-\lvert{m}\rvert)!}{(n+\lvert{m}\rvert)!}\int_0^\pi sin(\theta)d\theta \int_0^{2\pi}d\phi \times P^{\lvert{m}\rvert}_{n}[cos(\theta)]e^{im\phi} cB^{beam}_{rad}(r,\theta,\phi)$$

$$g^{m}_{n,TM} = \frac{(-i)^{n-1}}{2\pi}\frac{kr}{j_{n}(kr)}\frac{n-\lvert{m}\rvert)!}{(n+\lvert{m}\rvert)!}\int_0^\pi sin(\theta)d\theta \int_0^{2\pi}d\phi \times P^{\lvert{m}\rvert}_{n}[cos(\theta)]e^{im\phi} cE^{beam}_{rad}(r,\theta,\phi)$$

Where $B^{beam}_{rad}$ and $E^{beam}_{rad}$ are teh radial componetents of the beam field in sphereical coordiantes. If the beam is on-axis with respect to the sphere, only the $m=\pm1$ beam shape coefficents are non-zero and are given by,

$$g^{1}_{n,TM} = -ig_{n,TE}$$
$$g^{-1}_{n,TM} = ig_{n,TE}$$
$$g^{\pm1}_{n,TM} = g_{n,TE}$$

This process is called "obataining the beam shape coefficents by quadratures. Since the beam is assumed to be an exact solution of Maxwell's equations, the radial dependance of the angular integrals exactly cancels out the prefactory $kr/j_n(kr)$, and the resulting coefficents are constants. 

For the scattering of a Maxwell compliant Guassian beam by a sphere, the Mie discovered that scattered and interior Bromich potentials are,

$$U^{scatt}_{TE}(r,\theta,\phi) = \sum_{n=1}^\infty \sum_{m=-n}^n (\frac{E_0}{k})i^nC^{pw}_{n}B^{m}_{n}\zeta^{1}_nP^{\lvert{m}\rvert}_{n}[cos(\theta)]e^{im\phi}$$

$$U^{scatt}_{TM}(r,\theta,\phi) = \sum_{n=1}^\infty \sum_{m=-n}^n (\frac{E_0}{k})i^nC^{pw}_{n}A^{m}_{n}\zeta^{1}_nP^{\lvert{m}\rvert}_{n}[cos(\theta)]e^{im\phi}$$

and 

$$U^{interior}_{TE}(r,\theta,\phi) = \sum_{n=1}^\infty \sum_{m=-n}^n (\frac{NE_0}{k})i^nC^{pw}_{n}D^{m}_{n}\psi_n(Nkr)P^{\lvert{m}\rvert}_{n}[cos(\theta)]e^{im\phi}$$

$$U^{interior}_{TM}(r,\theta,\phi) = \sum_{n=1}^\infty \sum_{m=-n}^n (\frac{NE_0}{k})i^nC^{pw}_{n}C^{m}_{n}\psi_n(NKr)P^{\lvert{m}\rvert}_{n}[cos(\theta)]e^{im\phi}$$

where $\zeta^{1}_n$ are Riccati-Hankel functions, $A_n^m$ and $A_n^m$ are teh partial wave scattering amplitudes, and $C_n^m$ and $B_n^m$ are the partial wave interior amplitudes. Solving for boundry conditions gives,

$A_n^m = a_ng^m_{n,TM}$

$B_n^m = b_ng^m_{n,TE}$

$C_n^m = c_ng^m_{n,TM}$

$D_n^m = d_ng^m_{n,TM}$

### Scattering efficiency of transparent spheres

```python
#Plot the scattering efficiency as a function of wavelength for 4 micron glass spheres

radius=2
lam=np.linspace(0.2,1.2,200)
x=2*np.pi*radius/lam

# from https://refractiveindex.info/?shelf=glass&book=BK7&page=SCHOTT

#values determined from dispersion coeffiences
m2 = 1+1.03961212/(1-0.00600069867/lam**2)
m2 += 0.231792344/(1-0.0200179144/lam**2)
m2 += 1.01046945/(1-103.560653/lam**2)
m = np.sqrt(m2)

#set function to allow for various calculations
qext, qsca, qback, g = miepython.mie(m,x)

plt.plot(lam*1000,qsca)
plt.title("BK7 glass spheres 4 micron diameter")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Scattering Efficiency (-)")
```

```python
#Plot the scattering cross section as a function of wavelength for 2 micron water droplets

radius = 0.5 # in microns
lam = np.linspace(0.2,1.2,100) # also in microns
x = 2*np.pi*radius/lam

# from https://refractiveindex.info/?shelf=main&book=H2O&page=Daimon-24.0C
m2 = 1+5.666959820E-1/(1-5.084151894E-3/lam**2)
m2 += 1.731900098E-1/(1-1.818488474E-2/lam**2)
m2 += 2.095951857E-2/(1-2.625439472E-2/lam**2)
m2 += 1.125228406E-1/(1-1.073842352E1/lam**2)
m = np.sqrt(m2)
qext, qsca, qback, g = miepython.mie(m,x)
plt.plot(lam*1000, qsca)
plt.title(r"Water Droplets (1 μm diameter)")
plt.xlabel("Wavelength (nm)")
plt.ylabel(r"Scattering Cross Section (μm$^2$)")
```

### Scattering efficiency of small Gold Spheres

```python
#Use pandas to read CSV file
import pandas as pd
data=pd.read_csv("Ref_ind_Au.csv")

#read column values and write as np.arrays
ref_lam=data['Wavelength, µm'].to_numpy()
ref_n=data['n'].to_numpy()
ref_k=data['k'].to_numpy()

#Imput values for light dispersion
radius = 0.1 # in microns
m = ref_n-1.0j*ref_k   
x = 2*np.pi*radius/ref_lam
cross_section_area = np.pi * radius**2
mu_a = 4 * np.pi * ref_k / ref_lam # nm
qext, qsca, qback, g = miepython.mie(m,x)

#define scattering and absorption cross section
sca_cross_section = qsca * cross_section_area
abs_cross_section = (qext-qsca) * cross_section_area

#Create 3 subplots
plt.subplots(3,1,figsize=(9,9))

#First plot of real and imaginary refractive index vs. wavelength
plt.subplot(311)
plt.plot(ref_lam*1000, ref_n, 'ob')
plt.plot(ref_lam*1000, -ref_k, 'sr')
plt.title("Gold Spheres 100nm diameter")
plt.xticks([])
plt.text(700, 1, "real refractive index", color='blue')
plt.text(1100, -6, "imaginary refractive index", color='red')

#Second plot of 'absorption' depth(nm) vs. wavelength
plt.subplot(312)
plt.plot(ref_lam*1000, 1000/mu_a, 'ob')
plt.ylabel("Absorption Depth [nm]")

#Third plot of Absorption and Scattering vs. Wavelength
plt.subplot(313)
plt.plot(ref_lam*1000, abs_cross_section, 'ob')
plt.plot(ref_lam*1000, sca_cross_section, 'sr')
plt.xlabel("Wavelength (nm)")
plt.ylabel(r"Cross Section (μm$^2$)")
plt.text(700, 0.01, "absorption", color='blue')
plt.text(750, 0.1, "scattering", color='red')
plt.savefig("04_plot.png")
```

### Unpolarized Scattering Function

If unpolarized light hits the sphere, then there are no polarization effects to worry about.  It is possible to generate a plot to show how scattering changes with angle.

```python
m = 1.5 #magnitude of refractive index mismatch between particle and medium
x = np.pi/3
theta = np.linspace(-180,180,180)
mu = np.cos(theta/180*np.pi)
scat = miepython.i_unpolarized(m,x,mu)

fig,ax = plt.subplots(1,2,figsize=(12,5))
ax=plt.subplot(121, projection='polar')
ax.plot(theta/180*np.pi,scat)
ax.set_rticks([0.05, 0.1,0.15]) 
ax.set_title("m=1.5, Sphere Diameter = $\lambda$/3")

plt.subplot(122)
plt.plot(theta,scat)
plt.xlabel('Exit Angle [degrees]')
plt.ylabel('Unpolarized Scattered light [1/sr]')
plt.title('m=1.5, Sphere Diameter = $\lambda$/3')
plt.ylim(0.00,0.2)

plt.show()
```

### Rayleigh Scattering

Classic Rayleigh scattering treats small particles with natural (unpolarized) light.  
The solid black line denotes the total scattered intensity. The red dashed line is light scattered that is polarized perpendicular to the plane of the graph and the blue dotted line is for light parallel to the plane of the graph.

```python
m = 1.3
x = 0.01
theta = np.linspace(-180,180,180)
mu = np.cos(theta/180*np.pi)
ipar = miepython.i_par(m,x,mu)/2
iper = miepython.i_per(m,x,mu)/2
iun  = miepython.i_unpolarized(m,x,mu)

fig,ax = plt.subplots(1,2,figsize=(12,5))
ax=plt.subplot(121, projection='polar')
ax.plot(theta/180*np.pi,iper,'r--')
ax.plot(theta/180*np.pi,ipar,'b:')
ax.plot(theta/180*np.pi,iun,'k')

ax.set_rticks([0.05, 0.1,0.15]) 
plt.title('m=%.2f, Sphere Parameter = %.2f' %(m,x))

plt.subplot(122)
plt.plot(theta,iper,'r--')
plt.plot(theta,ipar,'b:')
plt.plot(theta,iun,'k')

plt.xlabel('Exit Angle [degrees]')
plt.ylabel('Normalized Scattered light [1/sr]')
plt.title('m=%.2f, Sphere Parameter = %.2f' %(m,x))
plt.ylim(0.00,0.125)
plt.text(130,0.02,r"$0.5I_{per}$",color="blue", fontsize=16)
plt.text(120,0.062,r"$0.5I_{par}$",color="red", fontsize=16)
plt.text(30,0.11,r"$I_{unpolarized}$",color="black", fontsize=16)


plt.show()
```
