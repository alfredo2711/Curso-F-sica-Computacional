# Matplotlib (continuación)

## animaciones

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,t):
    return np.sin(x-3*t)
```

```python
x=np.linspace(0, 10*np.pi, 1000)
```

```python
t=np.arange(0,4,0.1) #segundos
```

```python
xx, tt = np.meshgrid(x,t)
```

```python
F=f(xx, tt)
```

```python
# Si buscamos la función para t=0
```

```python
F[0]
```

```python
F[1]
```

```python
plt.plot(x, F[0], 'r')
plt.plot(x, F[5], 'b')
```

```python
for i in range(0, 10):
    plt.plot(F[i])
    plt.savefig("función{}.png".format(i))
    plt.show()
```

```python
import glob
from PIL import Image

fp_in="función*.png"
fp_out="animación.gif"

imgs=(Image.open(f) for f in sorted(glob.glob(fp_in)))
img=next(imgs) # se extrae la primera imagen de a iteración
img.save(fp=fp_out, format='GIF', append_images=imgs, save_all=True, duration=500, loop=0)
```

```python
#Otra forma
from matplotlib import animation
fig, ax=plt.subplots(1, 1)
b,=ax.plot([],[])
time_text=ax.text(0.65, 0.95,'',fontsize=15, bbox=dict(facecolor='white', edgecolor='black')
                 , transform=ax.transAxes)
ax.set_xlim(0,10*np.pi)
ax.set_ylim(-1.5, 1.5)

def animar(i):
    b.set_data(x,F[i])
    time_text.set_text('t={} s'.format(i/10))

ani=animation.FuncAnimation(fig, animar, frames=40)
ani.save('animación.gif', writer='pillow', fps=50, dpi=100)
```

```python
x=np.linspace(-1,1,100)
y=np.linspace(-1,1,100)
xx, yy= np.meshgrid(x,y)
z=xx**2 + yy**2 
```

```python
fig, ax=plt.subplots(subplot_kw={"projection":"3d"})
ax.plot_surface(xx,yy,z, cmap=plt.cm.hot)

ax.view_init(elev=10, azim=0)

def animar(i):
    ax.view_init(elev=10, azim=3*i)
    
ani=animation.FuncAnimation(fig, animar, frames=120)
ani.save('animación_3D.gif', writer='pillow', fps=3, dpi=100)
```
