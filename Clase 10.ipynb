# Matplotlib (continuación)

## Gráficas multidimensionales

Función contour

```python
import numpy as np
import matplotlib.pyplot as plt
```

$$f(x,y)=\cos{x}+\sin^2{y}$$

```python
def f(x,y):
    return np.cos(x)+np.sin(y)**2
```

```python
x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
xx, yy = np.meshgrid(x,y)
```

```python
plt.contour(xx, yy, f(xx, yy))
plt.text(-0.5,1.5,'hola', fontsize=20)
plt.colorbar()
```

```python
a=plt.contour(xx, yy, f(xx, yy))
plt.clabel(a)
plt.colorbar()
```

```python
plt.contourf(xx, yy, f(xx, yy), levels=100)
plt.colorbar()
```

```python
plt.contourf(xx, yy, f(xx, yy), np.linspace(-5,2,100))
plt.colorbar()
```

```python
plt.contourf(xx, yy, f(xx, yy), np.linspace(-5,2,100), cmap=plt.cm.hot)
plt.colorbar()
```

```python
plt.contourf(xx, yy, f(xx, yy), vmin=0.5, levels=100, cmap=plt.cm.hot)
plt.colorbar()
```

## Interfaz orientada a objetos (forma recomendada de graficar)

```python
def f(x):
    return np.exp(-x**2)
```

```python
x=np.linspace(-1,5,100)
```

```python
fig, axes = plt.subplots(1,1)
axes.plot(x,f(x),'ro', label='Función f(x)')
axes.set_xlabel('Ejex')
axes.set_xlim(0,2)
axes.legend()
axes.text(1,1, 'hola', transform=axes.transAxes, bbox=dict(facecolor='white', edgecolor='black'))
```

```python
fig, axes = plt.subplots(1,2, sharey=True)
axes[0].plot(x,f(x),'ro', label='Función f(x)')
axes[1].plot(x,-f(x),'go', label='Función -f(x)')
axes[0].legend()
axes[1].legend()
```

## Gráfica de vectores

```python
x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
xx, yy = np.meshgrid(x, y)
```

```python
# El vector tiene dos componentes (Ex, Ey)
#Ex (x, y)
#Ey (x, y)
```

```python
Ex=-1 - xx**2 + yy
Ey=1+xx-yy**2
```

```python
magnitud=np.sqrt(Ex**2 + Ey**2)
```

```python
fig, axes=plt.subplots(1,3, figsize=(15,5))
axes[0].streamplot(xx, yy, Ex, Ey)
axes[1].streamplot(xx, yy, Ex, Ey, color=magnitud, cmap=plt.cm.hot)
axes[2].streamplot(xx, yy, Ex, Ey, linewidth=7*magnitud/magnitud.max())
```

## Gráficos en 3D

```python
x=np.linspace(-1,1,100)
y=np.linspace(-1,1,100)
```

```python
xx, yy =np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
```

```python
fig, axes= plt.subplots(subplot_kw={"projection":"3d"})
axes.plot_surface(xx, yy,z, cmap=plt.cm.hot)
```
