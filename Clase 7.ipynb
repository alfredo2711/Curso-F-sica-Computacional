## Funciones en 2D

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

Graficar la función $z=\frac{sin(r)}{r}$, donde $r=\sqrt{x^2+y^2}$

```python
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
```

```python
x
```

```python
y
```

```python
np.meshgrid(x,y)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
r=np.sqrt(xx**2+yy**2)
```

```python
z=np.sin(r)/r
```

```python
plt.contour(xx,yy,z, levels=10)
```

```python
a=plt.contour(xx, yy, z, levels=10)
plt.clabel(a)
```

```python
plt.contourf(xx,yy,z, levels=50)
plt.colorbar()
```

```python
from matplotlib import cm
```

```python
plt.contourf(xx,yy,z, cmap=cm.hot, levels=100)
plt.colorbar()
```

## Ejemplo de funciones en 2D

Sea $f(x,y) = e^{-(x^2+y^2)} \cdot \sin(x)$ para $-2 \leq x \leq 2$ y $-2 \leq y \leq 2$

1. Haga un gráfico de contorno de $f$
2. Trace el perfil del gráfico para $y$=0
3. Encuentre el volumen bajo la curva de $|f(x,y)|$ en el rango dado para $x$ y $y$
4. Encuentre el volumen bajo la curva de $|f(x,y)|$ solo en el rango donde $\sqrt{x^2+y^2}>0.5$ 

```python
#1
x=np.linspace(-2,2,1000)
y=np.linspace(-2,2,1000)
xx, yy=np.meshgrid(x,y)
f=np.exp(-xx**2-yy**2)*np.sin(x)
```

```python
plt.contourf(xx,yy,f, levels=100)
plt.colorbar()
```

```python
#2
f2=np.exp(-x**2)*np.sin(x)
```

```python
plt.plot(x,f2)
```

```python
#3
abs(f).sum()*(x[1]-x[0])*(y[1]-y[0])
```

```python
#4
abs(f[np.sqrt(x**2+y**2)>0.5]).sum()*(x[1]-x[0])*(y[1]-y[0])
```

## Sistemas de ecuaciones

$3x+2y+z$=4

$5x-5y+4z$=3

$6x+z=0$

```python
A=np.array([[3,2,1],
          [5,-5,4],
          [6,0,1]])
```

```python
C=np.array([4,3,0])
```

```python
np.linalg.solve(A,C)
```
