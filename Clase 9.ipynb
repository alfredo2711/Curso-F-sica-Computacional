```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.loadtxt(fname='temperatura_1.csv', delimiter=',')
```

```python
a.shape
```

```python
#Las columnas representan mediciones de temperatura por día
#Las filas representan mediciones de temperatura se tomaban cada hora
```

```python
#Promedio en el primer día
a[:,0].mean()
```

```python
#Promedio en el segundo día
a[:,1].mean()
```

```python
#Promedio todos los días
b=a.mean(axis=0)
b
```

```python
#Graficar promedio por día
día=np.arange(1,len(b)+1)
plt.plot(día, b)
```

```python
#importar varios archivos a la vez
import glob
```

```python
files=glob.glob('temperatura*.csv')
```

```python
files
```

```python
#temperatura_1, temperatura_2, temperatura_3 se han tomado en sitios diferentes
```

```python
#Graficar el promedio por día en los tres diferentes puntos
```

```python
for i, f in enumerate(files):
    data=np.loadtxt(fname=f, delimiter=',')
    plt.plot(día, data.mean(axis=0), label='temperatura_{}'.format(i+1))
    plt.xlabel('día')
    plt.ylabel('promedio')
    plt.legend()
```

$$f(x)=e^{-x^2}$$

```python
x=np.linspace(-1,5)
```

```python
def f(x):
    return np.exp(-x**2)
```

```python
#Graficando una función
plt.figure(figsize=(10,5))
plt.plot(x, f(x), 'o-r', label="Función f(x)", lw=0.5, ms=5)
plt.xlabel("Eje $x$", fontsize=20)
plt.ylabel("$f(x)$", fontsize=20)
plt.tick_params(axis='both', labelsize=20)
plt.grid()
plt.legend(loc='upper left',fontsize=20)
plt.title("Función $f(x)$", fontsize=20)
```

```python
#Graficando dos funciones
```

```python
plt.plot(x, f(x), 'ro', label="Función f(x)")
plt.plot(x, 1-f(x), 'g--', label="Función 1- f(x)")
plt.ylim(top=1.5)
plt.legend(loc='upper right', ncol=2)
plt.show()
```

## Gráficas multidimensionales

```python
#Función scatter
x=np.random.randn(100)
y=np.random.randn(100)
```

```python
plt.scatter(x,y)
```

```python
#Añadimos una variable adicional cambiando el tamaño de las bolitas
z=50*np.random.randn(100)
plt.scatter(x,y, s=z)
```

```python
#Añadimos una variable adicional cambiando el color de las bolitas
temp=np.random.randn(100)
plt.scatter(x,y, s=z, c=temp, cmap=plt.cm.hot)
plt.colorbar(label='temperatura')
```
